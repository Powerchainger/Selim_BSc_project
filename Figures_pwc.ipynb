Imports

```python
import notebook
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from platform import python_version
import xgboost as xgb
import graphviz
import sklearn
from sklearn.metrics import mean_absolute_error as MAE
from sklearn.metrics import mean_absolute_percentage_error as MAPE
from statsmodels.graphics.tsaplots import plot_acf

print(f"python: v {python_version()}")
print(f"Jupyter Notebook: v {notebook.__version__}")
print(f"numpy: v {np.__version__}")
print(f"pandas: v {pd.__version__}")
print(f"seaborn: v {sns.__version__}")
print(f"graphviz: v {graphviz.__version__}")
print(f"matplotlib: v {matplotlib.__version__}")
print(f"sklearn: v {sklearn.__version__}")
print(f"XGBoost: v {xgb.__version__}")
```

Import Data

```python
mapping = {
    "kevin": 1,
    "matthijs": 2,
    "manolis": 6,
    "jet": 8,
    "robin": 9
    }
df_all = pd.read_csv('Predictions/Pwc_12689_complete.csv')
df_all_nilm = pd.read_csv('Predictions/Pwc_12689_nilm_complete.csv')
df_complete = pd.merge(df_all, df_all_nilm, on=['House', 'Date'], suffixes=('', '_nilm'))
df_complete['House'] = df_complete['House'].map(mapping)

print("CSV read successfully.")
```

Compute and display metrics and plots for each house

```python
df_complete['Date'] = pd.to_datetime(df_complete['Date'])
df_complete['Hour'] = df_complete['Date'].dt.hour
df_complete["Weekday"] = df_complete['Date'].dt.weekday

for house in mapping.keys():
    
    df_house = df_complete[df_complete['House'] == mapping[house]]
    xgb_rmse = np.sqrt(np.mean((df_house['XGB'] - df_house['Volume'])**2))
    xgb_mae = MAE(df_house['XGB'], df_house['Volume'])
    xgb_mape = MAPE(df_house['XGB'], df_house['Volume'])
    xgb_rmse_nilm = np.sqrt(np.mean((df_house['XGB_nilm'] - df_house['Volume'])**2))
    xgb_mae_nilm = MAE(df_house['XGB_nilm'], df_house['Volume'])
    xgb_mape_nilm = MAPE(df_house['XGB_nilm'], df_house['Volume'])
    gbm_rmse = np.sqrt(np.mean((df_house['GBM'] - df_house['Volume'])**2))
    gbm_mae = MAE(df_house['GBM'], df_house['Volume'])
    gbm_mape = MAPE(df_house['GBM'], df_house['Volume'])
    gbm_rmse_nilm = np.sqrt(np.mean((df_house['GBM_nilm'] - df_house['Volume'])**2))
    gbm_mae_nilm = MAE(df_house['GBM_nilm'], df_house['Volume'])
    gbm_mape_nilm = MAPE(df_house['GBM_nilm'], df_house['Volume'])
    rf_rmse = np.sqrt(np.mean((df_house['RF'] - df_house['Volume'])**2))
    rf_mae = MAE(df_house['RF'], df_house['Volume'])
    rf_mape = MAPE(df_house['RF'], df_house['Volume'])
    rf_rmse_nilm = np.sqrt(np.mean((df_house['RF_nilm'] - df_house['Volume'])**2))
    rf_mae_nilm = MAE(df_house['RF_nilm'], df_house['Volume'])
    rf_mape_nilm = MAPE(df_house['RF_nilm'], df_house['Volume'])
    combined_rmse = np.sqrt(np.mean((df_house['Predicted'] - df_house['Volume'])**2))
    combined_mae = MAE(df_house['Predicted'], df_house['Volume'])
    combined_mape = MAPE(df_house['Predicted'], df_house['Volume'])
    combined_rmse_nilm = np.sqrt(np.mean((df_house['Predicted_nilm'] - df_house['Volume'])**2))
    combined_mae_nilm = MAE(df_house['Predicted_nilm'], df_house['Volume'])
    combined_mape_nilm = MAPE(df_house['Predicted_nilm'], df_house['Volume'])
    
    print(f"House {mapping[house]} : ")
    print(f"XGBoost : RMSE: {xgb_rmse:.2f}, MAE: {xgb_mae:.2f}, MAPE: {xgb_mape:.2%}")
    print(f"XGBoost - NILM : RMSE: {xgb_rmse_nilm:.2f}, MAE: {xgb_mae_nilm:.2f}, MAPE: {xgb_mape_nilm:.2%}")
    print(f"GBM - RMSE: {gbm_rmse:.2f}, MAE: {gbm_mae:.2f}, MAPE: {gbm_mape:.2%}")
    print(f"GBM - NILM : RMSE: {gbm_rmse_nilm:.2f}, MAE: {gbm_mae_nilm:.2f}, MAPE: {gbm_mape_nilm:.2%}")
    print(f"RF - RMSE: {rf_rmse:.2f}, MAE: {rf_mae:.2f}, MAPE: {rf_mape:.2%}")
    print(f"RF - NILM : RMSE: {rf_rmse_nilm:.2f}, MAE: {rf_mae_nilm:.2f}, MAPE: {rf_mape_nilm:.2%}")
    print(f"Combined - RMSE: {combined_rmse:.2f}, MAE: {combined_mae:.2f}, MAPE: {combined_mape:.2%}")
    print(f"Combined - NILM : RMSE: {combined_rmse_nilm:.2f}, MAE: {combined_mae_nilm:.2f}, MAPE: {combined_mape_nilm:.2%}")


    plt.figure(figsize=(12, 6))
    plt.plot(df_house['Date'], df_house['Volume'], label='Actual', color='blue')
    plt.plot(df_house['Date'], df_house['Predicted'], label='Predicted', color='orange')
    plt.plot(df_house['Date'], df_house['Predicted_nilm'], label='Predicted NILM', color='green') 
    plt.xlabel('Date') 
    plt.ylabel('Energy Consumption (Watts)') 
    plt.title(f'House {mapping[house]} - Actual vs Predicted Energy Consumption') 
    plt.legend() 
    plt.grid(True) 
    plt.show()

    plt.figure(figsize=(12, 6))
    plt.plot(df_house['Date'], df_house['Residual'], label='Residuals', color='red') 
    plt.plot(df_house['Date'], df_house['Residual_nilm'], label='Residuals NILM', color='green')
    plt.axhline(0, color='red', linestyle='--', linewidth=1)
    plt.xlabel('Date') 
    plt.ylabel('Residual (Predicted - Actual)') 
    plt.title(f'House {mapping[house]} - Residuals Over Time') 
    plt.legend() 
    plt.grid(True) 
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.histplot(df_house['Residual'], label='Without Device information', color='red', kde=True)
    sns.histplot(df_house['Residual_nilm'], label='With Device information', color='blue', kde=True)
    plt.xlabel('Residual in Wh')
    plt.ylabel('Frequency')
    plt.title(f'Residual Distribution (House {mapping[house]})')
    plt.legend()
    # plt.tight_layout()
    plt.show()

    plt.figure(figsize=(12, 6))
    df_house.groupby('Weekday')['Residual'].mean().plot(kind='bar', label='Mean Residual by Weekday (No Device information)', color='red')
    df_house.groupby('Weekday')['Residual_nilm'].mean().plot(kind='bar', label='Mean Residual by Weekday (With Device information)', color='blue', alpha=0.7)
    plt.xlabel('Weekday')
    plt.ylabel('Mean Residual in Wh')
    plt.title(f'Mean Residual by Weekday (House {mapping[house]}) - Combined')
    plt.xticks(ticks=range(7), labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
    plt.legend()
    # plt.tight_layout()
    plt.show()

    plt.figure(figsize=(12, 6)) 
    df_house.groupby('Hour')['Residual'].mean().plot(kind='bar', label='Without Device information', color='red') 
    df_house.groupby('Hour')['Residual_nilm'].mean().plot(kind='bar', label='With Device information', color='blue', alpha=0.7) 
    plt.xlabel('Hour of Day (0-23)')
    plt.ylabel('Mean Residual in Wh')
    plt.title(f'Mean Residual by Hour (House {mapping[house]}) - Combined')
    plt.xticks(ticks=range(24), labels=[f'{h}:00' for h in range(24)], rotation=45)
    plt.legend()
    # plt.tight_layout()
    plt.show()

    #  plot acf of residuals
    max_lags = 24 * 7
    if len(df_house) < max_lags:
        max_lags = len(df_house) - 1
    plt.figure(figsize=(10, 4))
    plot_acf(df_house['Residual'], lags=max_lags, ax=plt.gca())
    plt.title(f'ACF plot of Residuals (House {mapping[house]}) - Combined - Without Device information')
    plt.xlabel('Lag')
    # plt.ylabel('Autocorrelation')
    plt.tight_layout()
    # plt.grid(True)
    plt.show()

    plt.figure(figsize=(10, 4))
    plot_acf(df_house['Residual_nilm'], lags=max_lags, ax=plt.gca())
    plt.title(f'ACF plot of Residuals (House {mapping[house]}) - Combined - With Device information')
    plt.xlabel('Lag')
    # plt.ylabel('Autocorrelation')
    plt.tight_layout()
    # plt.grid(True)
    plt.show()  
```
